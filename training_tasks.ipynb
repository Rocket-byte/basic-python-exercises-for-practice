```python
def year_is_high (year:int) -> str:
  if (year%4 == 0 and year%100 != 0) or year%400 == 0:
    return "yes"
  else:
    return "no"

x = int(input("Enter the year: "))

print(year_is_high(x))
```

```python
def is_same_color (x:int,y:int,m:int,n:int) -> str:
  first_cell = x+y
  second_cell = m+n
  if first_cell%2 == second_cell%2:
    return "yes"
  else:
    return "no"

x = int(input("Enter first number for the first cell: "))
y = int(input("Enter second number for the first cell: "))
m = int(input("Enter first number for the second cell: "))
n = int(input("Enter second number for the second cell: "))

print(is_same_color(x,y,m,n))

```

```python
def lesson_finish_at (x:int) -> tuple:
  lesson_time_start = 9*60        # 540 min
  lesson_duration_min = 45*x
  lesson_break_time = 0

  for i in range(1,x):
    if i%2 == 1:
      lesson_break_time += 5
    else:
      lesson_break_time += 15

  total_minutes = lesson_time_start + lesson_duration_min + lesson_break_time
  hours = total_minutes//60
  minutes = total_minutes%60

  return hours,minutes


x = int(input("Enter number of lesson: "))
h, m = lesson_finish_at(x)
print(f"Lesson finish at {h}:{m}")
```

```python
def n_factorail (x:int) -> int:
  res = 1

  for i in range(1, x+1):
    res *= i

  return res

x = int(input("Enter number: "))
print(n_factorail(x))
```

```python
def sum_factorials (x:int) -> int:
  fact = 1
  sum = 0

  for i in range (1, x+1):
    fact *= i
    sum += fact

  return sum

x = int(input("Enter number: "))
print(sum_factorials(x))
```

```python
def replace_words (x:str) -> str:
  words = x.split()
  reversed_x = " ".join(words[::-1])

  return(reversed_x)

x = input("Enter string: ")
print(replace_words(x))
```

```python
def min_div (x:int) -> int:
  for i in range(2,x+1):
    if x%i == 0:
      return i

x = int(input("Enter number higher than 2: "))
print(min_div(x))
```

```python
def count_increases () -> int:
  prev = int(input())
  count = 0

  while True:
    curr = int(input())
    if curr == 0:
     break

    if curr > prev:
      count += 1

    prev = curr

  return count

print(count_increases())
```

```python
def count_local_maxima (x:list[int]) -> int:
 count = 0

 for i in range(1, len(x)-1):
    if x[i] > x[i-1] and x[i] > x[i+1]:
      count += 1

 return count


nums = [1,3,2,4,1]
print(count_local_maxima(nums))
```

```python
def find_duplicates(nums: list[int]) -> list[int]:
  duplicates = [] # list
  for i in nums:
    if nums.count(i) > 1:
      duplicates.append(i)
  return duplicates


nums = [1,2,2,3,3]
print(find_duplicates(nums))
```
